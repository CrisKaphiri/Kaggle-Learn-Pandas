### Lambda con map y apply

```python
# Se obtiene el promedio
reviews_points_mean = reviews['points'].mean()
# Se obtiene una nueva serie con los valores
reviews['points'].map(lambda p: p - reviews_points_mean)
```

```python
def remean_points(row):
    row["points"] = row["points"] - reviews_points_mean
    return row

# Obtiene un df con el formula aplicada, pero no se aplica al df original
reviews.apply(remean_points, axis = 1)
```

```python
reviews = pd.read_csv('./data/winemag-data-130k-v2.csv', index_col= 0)
reviews.head(8)
```

```python
# 1. Obtén la media de la columna 'points' del df
reviews['points'].median()
```

```python
# 3. ¿Con qué frecuencia aparece cada país en el df?
reviews['country'].value_counts()
```

```python
# 4. Crea una columna con el precio promedio restado
price_mean = reviews['price'].mean()
price_mean
```

```python
# Se obtiene una nueva serie con los valores
reviews['price'].map(lambda p: p - price_mean)
```

```python
# 5. ¿Cuál vino es la mejor ganga? 
# Crea una variable con el titulo del vino y la relacion puntos-precio mas alta en el df
# Obtiene el indice del mejor vino calidad-precio, el indice se obtiene con idxmax()
wine_point_to_price =  (reviews['points'] / reviews['price']).idxmax()
reviews.loc[wine_point_to_price, 'title']
```

```python
reviews.loc[reviews['title'] == 'Bandit NV Merlot (California)']
```

```python
# 6. Crea una serie contando cuantas veces aparece cada una de las palabras 'tropical' y 'fruity' en la columna descriptión (Ignorar mayusculas)
# Mapea descripción, contando y sumando cuantas veces se repite la palabra tropical
count_tropical = reviews['description'].map(lambda desc: 'tropical' in desc).sum()
count_tropical
```

```python
# Mapea descripción, contando y sumando cuantas veces se repite la palabra 'fruity'
count_fruity = reviews['description'].map(lambda desc:'fruity' in desc).sum()
count_fruity
```

```python
# Crea la serie con los valores anteriores
desc_tropical_fruity = pd.Series([count_tropical, count_fruity], index = ['Tropical', 'Fruity'])
desc_tropical_fruity
```

```python
# 7. Nos gustaría publicar una reseña de vinos en nuestro sitio web, pero el sistema de calificacion oscila entre 80 - 100 puntos y es dificil de entender.
# Crea una serie con la cantidad de estrellas correspondientes a cada reseña en el df, considerando lo siguiente:
# 3 estrellas para mayor o igual a 95
# 2 estrellas entre 85 y 94
# 1 estrella para menor a 85
# 3 estrellas para Canada

reviews['star rating'] = 0
reviews.loc[(reviews['points'] >= 95) | (reviews['country'] == 'Canada'), 'star rating'] = 3
reviews.loc[(reviews['points'] >= 85) & (reviews['points'] < 95), 'star rating'] = 2
reviews.loc[reviews['points'] < 85, 'star rating'] = 1

reviews['star rating']
```

```python
# Para comprobar vemos los valores únicos
reviews['star rating'].unique()
```
